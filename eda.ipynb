```python
def calcula_ganancia(a, b, c, d, e, f):
    return (b * c + d + e + f) - a
```

```python
# Si se vende el equipo a su Valor Residual en este momento, así se verían las ganancias.
df['Ganancia'] = df.apply(lambda row : calcula_ganancia(row.MONTO_INICIAL_ACTIVO_S_IVA, row.RENTA, row.PLAZO, row.Comisiones, row.RENTAS, row['Valor Residual']), axis = 1)
```

```python
df.Ganancia.plot()
```

```python
df[df.Ganancia < -1e+08]
```

```python
ganancias_df = df[df.Ganancia > 0]
```

```python
# Para calcular las ganancias totales generadas con cada cliente
# y así encontrar al mejor cliente y peor cliente

ganancias_por_cliente = df.groupby(['ID_CLIENTE']).Ganancia.sum()
```

```python
print(ganancias_por_cliente.idxmax(), ganancias_por_cliente.max())
```

```python
print(ganancias_por_cliente.idxmin(), ganancias_por_cliente.min())
```
